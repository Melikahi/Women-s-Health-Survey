Women's Health Survey

In 1985, the USDA commissioned a study of women’s nutrition. Nutrient intake was measured for a random sample of 737 women aged 25-50 years. The following variables were measured:

- Calcium(mg)
- Iron(mg)
- Protein(g)
- Vitamin A(μg)
- Vitamin C(mg)


# Descriptive Analysis

We are going to analyse the dataset on the basis of descriptive analysis functions that will help us in understanding the pattern and behaviour of the data.

```python
#import all the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load the dataset
data= pd.read_csv("/nutrient.csv")
```

```python
data.head(2)
```

```python
print("data.columns: ")
print(data.columns)

print("data.dtypes: ")
print (data.dtypes)

```

```python
data.describe()
```

```python
#Measure of Central Tendency
print("\nMean: ")
print(data.mean())

print("\nMedian: ")
print(data.median())

print("\nMode: ")
print(data.mode())
```

```python
#Measure of Dispersion
print("/nVarince: ")
print (data.var())

print("\nStandard deviation: ")
print(data.std())

print("\nRange: ")
```

```python
#skewness and kurtosis
print("\nskewness: ")
print(data.skew())

print("\nkurtosis: ")
print(data.kurtosis())
```

```python
#Graph Analysis: Bar Chart
plt.figure(figsize=(6,4))
plt.bar(data['id'], data['calcium'], color = 'red', label='calcium')
plt.xlabel("ID")
plt.ylabel("Calcium")
plt.title("Calcium Intake by ID")
plt.legend()
plt.show()
```

```python
#Graph Analysis: Histogram
plt.figure(figsize=(6,4))
plt.hist(data['calcium'], bins=10, color = 'yellow', edgecolor='red')
plt.xlabel("ID")
plt.ylabel("Calcium")
plt.title("Histogram of Calcium Intake by ID")
plt.legend()
plt.show()
```
